```python
from PIL import Image
```

```python
filename = 'images/colosseum_lr.png'
im = Image.open(filename)
```

```python
width, height = im.size
print(f'width = {width}  height = {height}')
```

```python
left = width * 37/64
top = height * 7/8
right = width * 21/32
bottom =  height 
 

im1 = im.crop((left, top, right, bottom))
```

```python
display(im1)
```

```python
im1.save('crop_' + filename)
```
